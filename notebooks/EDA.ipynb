```python
df['Date']=pd.to_datetime(df['Date'], errors='coerce')
```

```python
df.head()
```

```python
df['Close'] = df['Close'].astype(float)
df['Open'] = df['Open'].astype(float)
df['High'] = df['High'].astype(float)
df['Low'] = df['Low'].astype(float)
df['Volume'] = df['Volume'].astype(int)
```

```python
df['Year']=df['Date'].dt.year
df['Day']=df['Date'].dt.day
df['Month']=df['Date'].dt.month
```

Giá Đóng Cửa Từ Năm 2020 Đến 2025

```python
plt.figure(figsize=(12,6))
plt.plot(df['Date'],df['Close'],label='Close Price')
plt.title('Giá Đóng Cửa Của MSFT Từ Năm 2020 Đến 2025')
plt.xlabel('Năm'),plt.ylabel('Giá')
plt.legend()
plt.grid(True)
```

```python
years=df['Year'].unique()
for y in years:
    subset=df[df['Year'] == y]
    plt.figure(figsize=(12,4))
    plt.plot(subset['Date'],subset['Close'],label= 'Giá Đóng Cửa Theo Năm '+ str(y))
    plt.title('Giá Đóng Cửa Theo Năm'+str(y))
    plt.legend()
    plt.grid(True)
```

Đường Trung Bình Động Và Biến Động Dài Hạn

```python
df['SMA50']=df['Close'].rolling(50).mean()
df['SMA200']=df['Close'].rolling(200).mean()

plt.figure(figsize=(12,4))
plt.plot(df['Date'],df['Close'],label='Close Price')
plt.plot(df['Date'],df['SMA50'],label='SMA 50 ngày',color='orange')
plt.plot(df['Date'],df['SMA200'],label='SMA 200 ngày',color='red')
plt.title("MSFT với đường trung bình động dài hạn (5 năm)")
plt.legend()
plt.grid(True)
```

```python
df['Daily Returns']= df['Close'].pct_change()
plt.Figure(figsize=(12,4))
plt.plot(df['Date'],df['Daily Returns'],color='red')
plt.title('Lợi Nhuận Hàng Ngày của năm '+str(y))
plt.grid(True)
plt.legend()
plt.show()
```

```python
df['Volatility_20']=df['Daily Returns'].rolling(20).std()
plt.figure(figsize=(12,4))
plt.plot(df['Date'],df['Volatility_20'], color='orange')
plt.title('Biến động giá trong 20 ngày của MSFT')
plt.grid(True)
plt.legend()
plt.show()
```

```python
df['Volatility_20']
```

```python
years = df['Year'].unique()
for y in years:
    subset=df[df['Year']==y]
    plt.Figure(figsize=(12,4))
    plt.plot(subset['Date'],subset['Volume'], label='khối lượng giao dịch của năm' + str(y))
    plt.xlabel('Năm'),plt.ylabel('Khối Lượng Giao Dịch')
    plt.title('Xu hướng khối lượng giao dịch của năm ' + str(y))
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
years =df['Year'].unique()
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
for y in years:
    subset=df[df['Year']==y]

    mpf.plot(subset,type='candle',
             volume=True,
             title='Biểu Đồ Nến MSFT - Năm'+str(y),
             style='yahoo')
```

```python
df['Cumulative Return']=(1+df['Daily Returns']).cumprod()
plt.Figure(figsize=(12,4))
plt.plot(df['Cumulative Return'], color ='orange')
plt.title('Lợi Nhuận Quỹ Kế')
plt.grid(True)
plt.legend()
plt.show()
```
